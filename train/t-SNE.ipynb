```python
import keras
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

(X, Y) = keras.datasets.mnist.load_data()
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (10, 6)
```

```python
def sample_mnist(p):
    """
    grab a subsample of mnist where p is the proportion of the original set
    """
    SIZE = 60000
    rand = np.random.rand(SIZE) < p
    return (X[0][rand], X[1][rand])

images, labels = sample_mnist(.02)
train = images.reshape(images.shape[0], 28*28)
```

```python
model = TSNE(n_components = 2, 
             perplexity = 25,
             random_state = 0,
             n_iter = 1000,
             verbose = True)
tsne_train = model.fit_transform(train)
```

```python
plt.scatter(x = tsne_train[:, 0], 
            y = tsne_train[:, 1],
            alpha = 0.5,
            c = labels)
```
